# Prepare fasta from consensus regions

No terminal com o ambiente virtual newscenicplus aberto
* ``conda install -c bioconda bedtools``
* ``pip install flatbuffers``

```python
SCRIPT_DIR="./create_cisTarget_databases"
GENOME_FASTA="./data/hg38.fa"
CHROMSIZES="./data/hg38.chrom.sizes"
REGION_BED="../pycistopic/outs/consensus_peak_calling/consensus_regions.bed"
OUTPUT_FASTA="hg38.custom_database.with_1kb_bg_padding.fa"

bash "${SCRIPT_DIR}/create_fasta_with_padded_bg_from_bed.sh" \
        "${GENOME_FASTA}" \
        "${CHROMSIZES}" \
        "${REGION_BED}" \
        "${OUTPUT_FASTA}" \
        1000 \
        yes
```

```python
ls motif_colleciton/v10nr_clust_public/singletons > motifs.txt
```

```python
chmod +x cbust
export PATH=$PATH:$(pwd)
```

```python
OUT_DIR=""${PWD}""
CBDIR="${OUT_DIR}/motif_colleciton/v10nr_clust_public/singletons"
FASTA_FILE="${OUT_DIR}/hg38.custom_database.with_1kb_bg_padding.fa"
MOTIF_LIST="${OUT_DIR}/motifs.txt"

"${SCRIPT_DIR}/create_cistarget_motif_databases.py" \
    -f ${FASTA_FILE} \
    -M ${CBDIR} \
    -m ${MOTIF_LIST} \
    -o ${OUT_DIR}/${DATABASE_PREFIX} \
    --bgpadding 1000 \
    -t 8
```
